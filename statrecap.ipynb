```python
df2=pd.read_csv("HRDataset_v14.csv")
df2
```

```python
df2.columns
```

```python
df3=df2[["EmpID","DeptID","Department",'Salary']]
df3.head()
```

```python
df2["Department"].value_counts()
```

```python
df2.groupby(["Department"])["Salary"].mean()
```

```python
mean=df2.groupby("Department").Salary.mean().reset_index()
mean
```

```python
display(df2["Salary"].mode())
```

```python
display(mean["Salary"].var())
```

```python
median=df2.groupby("Department").Salary.median().reset_index()
display(median)
```

```python
display(median["Salary"].std())
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df2.groupby("Department")["Salary"].mean().sort_values().plot.barh(figsize=(10,5))
plt.xlabel("Salary")
plt.ylabel("Department")
plt.title("salary on dept")
plt.show()
```
