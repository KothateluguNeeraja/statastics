```python
df["FLOODS"]=df["FLOODS"].map({"YES":1,"NO":0})
```

```python
df.head()
```

```python
df["data_of_june"]=(df["JUN"]>=500).astype("int")
df["data_of_june"]
```

```python
df["data_of_jul"]=(df["JUL"]>=500).astype("int")
df["data_of_jul"]
```

```python
df.head()
```

```python
small_data=df.loc[:,["YEAR","data_of_june","data_of_jul","FLOODS"]]
small_data.head()
```

```python
#pd.crosstab()
pd.crosstab(small_data["FLOODS"],small_data["data_of_jul"])
```

```python
jul=small_data["data_of_jul"].value_counts().sort_values().plot(kind="bar")
jul
```

```python
sns.countplot(x="data_of_jul",hue="FLOODS",data=small_data)
```
